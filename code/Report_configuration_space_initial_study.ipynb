# Comparison of low- and high-energy configuration spaces for HMF systems

We want to check whether it is possible to identify specific differences in the topological structure of the configuration spaces of an HMF system when considered over below the transition point.  
To do this, from a sampling of a configuration space we will construct a metrical space and then consider its persistent homology, the expectation being that it will highlight a difference in the topological features.  

```python
import pandas as pd
import os, sys
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

## Initial Analysis

### Extraction and computation of the configuration space at low energy

```python
notebook_mode=True;

if notebook_mode==True:
    config_file='../data/low_energy/posit00.dat'
    output_dir='../data/low_energy/';
    config_tag='low_energy_HMF';
    num_coords=1000; #number of dof for each configuration point

else:
    config_file=sys.argv[1];
    output_dir=sys.argv[2];
    config_tag=sys.argv[3];
    num_coords=int(sys.argv[4]);

if not os.path.exists(output_dir):
    os.makedirs(output_dir);

f = open(config_file,'r');
ConfigSpace={};
for nu,line in enumerate(f.readlines()[0:]):
    # create points in conf space
    if nu/num_coords not in ConfigSpace:
        ConfigSpace[nu/num_coords]=[];
        if (nu/num_coords)%100==0:
            print 'point id ', nu/num_coords;
    # dumps coords into conf space point dict
    if line.rstrip('\n').strip()!='q':
        ConfigSpace[nu/num_coords].append(float(line));
    else:
        print 'Found error at line ', nu;
        ConfigSpace[nu/num_coords].append(float(0)); #sostituisco i missing points con 0 (non c'è ne sono in pratica)
    
lowCS = pd.DataFrame(ConfigSpace)
```

### Extraction and computation of the configuration space at high energy

```python
notebook_mode=True;

if notebook_mode==True:
    config_file='../data/high_energy/posit00.dat'
    output_dir='../data/high_energy/';
    config_tag='high_energy_HMF';
    num_coords=1000; #number of dof for each configuration point

else:
    config_file=sys.argv[1];
    output_dir=sys.argv[2];
    config_tag=sys.argv[3];
    num_coords=int(sys.argv[4]);

if not os.path.exists(output_dir):
    os.makedirs(output_dir);

f = open(config_file,'r');
ConfigSpace={};
for nu,line in enumerate(f.readlines()[0:]):
    # create points in conf space
    if nu/num_coords not in ConfigSpace:
        ConfigSpace[nu/num_coords]=[];
        if (nu/num_coords)%100==0:
            print 'point id ', nu/num_coords;

    # dumps coords into conf space point dict
    
    if line.rstrip('\n').strip()!='q':
        ConfigSpace[nu/num_coords].append(float(line));
    else:
        print 'Found error at line ', nu;
        ConfigSpace[nu/num_coords].append(float(0)); #sostituisco i missing points con 0 (check with authors)
    
highCS = pd.DataFrame(ConfigSpace)
```

Before proceeding it is interesting to take a look even just visually at the two configurations spaces obtained. In the plots below we show the values of the degrees of freedom (the phases) for each node at the various points in the two configuration spaces. In particular, columns correspond to points in conf. space, while rows to the individual spins.  
 
It is easy to see visually that, in the low energy regime, phases seem to display a higher coherence then in the high energy regime. However, in the low energy case we see a very different behaviour between roughly the first and the second halves of the simulated points. In the second part, spins lock onto a well defined average phase with extremely small dispersion around this value. I don't know what this is supposed to mean, but to me it looks like the system hit its ground state or at least some very long-lived metastable state.

## Mean-std scatter analysis

Before throwing the whole persistent homology machinery to these data, it might be interesting to compare the differences obtained already at this level in terms of the global properties of the system.  
In particular, we can project the observed points in configuration space into a much simpler 2d space, labeled by their average and standard deviation, obtaining a significantly simplified picture. It turns out that in the low energy regime the points lie on a well define curve in this space, while in the high energy regime instead the point cloud explores all points within a given region.

# Full topological analysis

**Now what happens if we instead we directly attack the data with persistent homology without previous refinement?**
  
We can embed the points in a number of different ways. Here we choose the simplest, that is a normal Euclidean distance between the vectors representing each configuration space point.  
For numerical reasons, we extract 100 points randomly and use those for the persistent homology calculation. Then, we repeat in a sort of boostrap. However, the results turn out to be rather consistent across different realizations, so we only show one for each energy regime (*I'm also running those obtained from max-min which should not require bootstrap*).  

```python
wlow_dict=pickle.load(open('../data/low_energy/weighed_gen_report.pck'))
whigh_dict=pickle.load(open('../data/high_energy/weighed_gen_report.pck'))
```

At the level of connected components ($H_0$), we do see some difference. In the low energy regime points start to connect to each other already at small differences, while for the high energy the persistence of the $H_0$ generators suggests that the points are almost all at the same distance from the others. Therefore, until the Rips parameter reaches that distance the components remain separated and rapidly connect completely the system when that distance is reached.  
Moreover, note also that the giant connected component appears around the same distance values (~80) in the two cases, the number of connected components starts to decrease much earlier for low energy (~20) as compared to the high energy regime (~65).  
For low energies this therefore seems to suggest that the configuration space points are clustered in an hierarchically organized fashion, giving rise to the observed patterns. This is also consistent with the *vertical bands* of similar configurations evident in Out[6] for low energy states.   


```python
ho.barcode_creator(wlow_dict[0],sizea=8,sizeb=8,W=1);
plt.title(r'Barcode for $H_0$ low energy',fontsize=20);

plt.show()
ho.barcode_creator(whigh_dict[0],sizea=8,sizeb=8,W=1);
plt.title(r'Barcode for $H_0$ high energy',fontsize=20);

plt.show()
```

In terms of $H_1$ instead the situation is significantly less exciting. We only see signals which are hardly distinguishable from topological noise appearing when the components are connecting to each other. Here we have therefore to conclude that at this level we cannot see significant differences in the higher homology patterns displayed by the configuration spaces themselves. 

```python
plt.figure(figsize=(20,8))
plt.subplot(121)
if len(wlow_dict[1])>1:
    plt.title(r'Persistence diagram for $H_1$ low energy case',fontsize=20);
    ho.complete_persistence_diagram(wlow_dict[1],W=1);
    plt.grid();
    plt.xlim(0,80)
    plt.ylim(0,80)
    plt.plot([0,100], [0,100],'--k')
#    plt.show()

plt.subplot(122)
if len(whigh_dict[1])>1:
    plt.title(r'Persistence diagram for $H_1$ high energy case',fontsize=20);
    ho.complete_persistence_diagram(whigh_dict[1],W=1);
    plt.grid();
    plt.xlim(0,80)
    plt.ylim(0,80)
    plt.plot([0,100], [0,100],'--k')
    plt.show()
```
